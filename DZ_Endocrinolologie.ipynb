```python
from scipy import stats

def histogram_m(df, var1, var2, bins):
  # Create a histogram using Seaborn
  ax = sns.histplot(data=df, x=var1, hue=var2, multiple="stack", bins=15)
  #plt.xlabel("Age (years)")
  plt.title(f"{var1} Distribution by {var2}")

  group, number, list_ ='group', 1, []

  for cat in df[var2].unique():
    group=group+str(number)
    #print(group)
    list_.append([group, cat, df[df[var2] == cat][var1]])
    group=group[:-1]
    number+=1


  _, p_value = stats.mannwhitneyu(list_[0][2], list_[1][2], alternative='two-sided')

  # Display the p-value on the plot
  plt.text(
      x=30,  # X-coordinate of the annotation (adjust as needed)
      y=140,   # Y-coordinate of the annotation (adjust as needed)
      s=f'Mann-Whitney U = {_} \n p-value: {p_value:.4f}',  # Text to display
      ha='center',   # Horizontal alignment
      va='bottom',   # Vertical alignment
      fontsize=10,   # Font size
      color='black'  # Text color
  )

  file_path = f'/content/drive/MyDrive/DZ_endocrinologie/{var1}-{var2}'
  plt.savefig(file_path, dpi=dpi)

  plt.show()

histogram_m(df, var1="Virsta, anii", var2="Sexul", bins=20)

```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import spearmanr


# subset - which var we need to consider to remove NA
# x - x axis var
# y - y axis var
# h - color for var
# col - the var for
def scatter_2(subset, x, y, h, col):
  # Filter out rows with missing values
  df_cleaned = df.dropna(subset=subset)

  # Create a relplot using Seaborn
  g = sns.relplot(x=x, y=y, hue=h, col=col, data=df_cleaned, height=3.5, kind='scatter')

# Calculate Spearman's correlation coefficient and 95% CI
  for ax in g.axes.flat:
    data_subset = df_cleaned[df_cleaned[col] == ax.get_title()[-1]]  # Filter data for each subplot
    correlation, p_value = spearmanr(data_subset[x], data_subset[y])

    # Bootstrap to calculate 95% CI for the correlation
    np.random.seed(0)  # Set a random seed for reproducibility
    n_iterations = 1000  # Number of bootstrap iterations
    bootstrap_correlations = []

    for _ in range(n_iterations):
        # Resample the data with replacement
        resampled_data = data_subset.sample(len(data_subset), replace=True)
        resampled_correlation, _ = spearmanr(resampled_data[x], resampled_data[y])
        bootstrap_correlations.append(resampled_correlation)

    # Calculate 95% CI
    lower_ci = np.percentile(bootstrap_correlations, 2.5)
    upper_ci = np.percentile(bootstrap_correlations, 97.5)

    # Annotate the correlation with 95% CI
    ax.annotate(f"Spearman's ρ = {correlation:.2f}\n95% CI: [{lower_ci:.2f}, {upper_ci:.2f}]",
                xy=(0.5, 0.9), xycoords='axes fraction',
                fontsize=10, color='black', ha='right')

    # Add an interpolation line with 95% CI using sns.regplot
    sns.regplot(x=x, y=y, data=data_subset, ax=ax, scatter=False, color='black', ci=95)

  plt.ylim(0, df[y].max()+df[y].max()*0.1)
  file_path = f'/content/drive/MyDrive/DZ_endocrinologie/{x}-{y}-{h}'
  plt.savefig(file_path, dpi=dpi)
  plt.show()

scatter_2(subset=['Virsta, anii', 'IMC'], x='Virsta, anii', y='IMC', h='Sexul', col='Sexul')
```
